## CSV

### Lectura de un csv

Para leer un archivo **csv**, lo primero tenermos que importar la librer√≠a que nos permite utilizar las funciones necesarias para su procesamiento, tanto de lectura como de escritura, esta librer√≠a es **csv**

Importamos la librer√≠a

```python
import csv
```

Salvo que estemos utilizando la librer√≠as como **pandas** que tienen otro tipo de funciones espec√≠ficas para la lectura de un archivo, con la librer√≠a csv, tendremos que utilizar la funci√≥n <code>open()</code> que recibir√° los siguientes par√°metros: El nombre del archivo csv y, el modo de apertura del archivo que por lo general, utilizaremos los siguientes:

* **r**: Lectura (*read*), abre un archivo para su lectura, devuelve un error si no existe.
* **w**: Escritura (*write*), se utiliza cuando vamos a escribir un archivo.

Adem√°s tambi√©n se puede especificar el tipo de archivo que se va a procesar:

* **t**: Texto.
* **b**: Binario, por ejemplo im√°genes.


M√°s informaci√≥n sobre la funci√≥n **open** https://www.w3schools.com/python/ref_func_open.asp

```python
compras_file = open('compras.csv', mode='rt')
```

Por si misma la asginaci√≥n a compras_file, no contienen nada, simplemente el objeto de la librer√≠a csv, que indica que hemos realizado la apertura del archivo 'compras.csv'

```python
compras_file
```

Para comenzar a leer el archivo, tenemos que crear el objeto *reader*, para ello utilizamos la func√≠on <code>csv.reader</code>

```python
reader_compras = csv.reader(compras_file)
```

De nuevo, a√∫n no tenemos nada, simplemente el objeto lector sobre el csv compras

```python
reader_compras
```

Para ver el contenido interior del mismo, tendr√≠amos o bien recorrer l√≠nea a l√≠nea el archivo o, utilizar una lista para guardar todo el archivo.

```python
clean_compras = list(reader_compras)

clean_compras
```

```python
compras_file = open('compras.csv', mode='rt')
reader_compras = csv.reader(compras_file)
archivo_compras = []

for linea in reader_compras:
    print('L√≠nea del archivo: ', linea)
    archivo_compras.append(linea)
```

```python
archivo_compras
```

Como podemos comprobar hay algunos fallos de escritura como 'Bol√É\xadgrafo Bic', para solventar estos problemas tenemos que utilizar una codificaci√≥n espec√≠fica del archivo, para esta tarea, utilizamos el par√°metro `encoding`, lo m√°s habitual es utilizar la codificaci√≥n **utf-8**

```python
# Modificamos la codificaci√≥n
compras_file = open('compras.csv', mode='rt', encoding='utf-8')

reader_compras = csv.reader(compras_file)
clean_compras = list(reader_compras)
clean_compras
```

## JSON

### Lectura de un archivo json

importamos la librer√≠a json

```python
import json
```

Utilizaremos el mismo procedimiento que con los archivos csv, se utiliza la funci√≥n `open`, para los archivos json, vamos a distinguir dos funciones principales para poder procesarlos, una es `json.load` y `json.loads`

La principal diferencia entre ambas funciones es que `json.load`se encarga de procesar un archivo json que est√© correctamente formateado como tal y tenga una estructura definida. Por su parte, `json.loads` se utiliza para cargar un json como un string, es decir, que no ha sido correctamente formateado, cuando estemos utilizando `json.loads` vamos a procesar el archivo **l√≠nea a l√≠nea**

```python
with open('mercados.json', 'r') as f:
    mercados = json.load(f)
```

```python
mercados
```

Una vez que cargamos el archivo con `json.load` vemos que es exactamente igual que un diccionario de datos y, podemos emplear las funciones que ya conocemos como keys o values

```python
mercados.keys()
```

```python
mercados['IBEX_MEDIUM_CAP']
```

```python
for company in mercados['IBEX_MEDIUM_CAP']:
    print(company)
```

```python
for company in mercados['IBEX_MEDIUM_CAP']:
    for value in company.keys():
        print(company[value])
```

Cuando el json no tiene formato... `json.loads`

```python
with open('tweets.txt', 'r') as f:
    tweets = json.load(f)
```

Al no tener formato, da error, ya que se trta de una composici√≥n de varios json el archivo.

```python
with open('tweets.txt', 'r') as f:
    tweets = json.loads(f)
```

Tampoco podemos procesar el archivo directamente con la funci√≥n `json.loads` ya que nos especifica que debe recibir una √∫nica cadena de texto, el archivo, son varias cadenas de texto.

Por lo tanto, tenemos que procesar el archivo **l√≠nea a l√≠nea**

```python
# Guardamos el json en una lista
lista_tweets = []

with open('tweets.txt', 'r') as f:
    # Recorremos l√≠nea a l√≠nea 
    for linea in f:
        tweets = json.loads(linea)
        lista_tweets.append(tweets)
```

## XML

### Lectura de un archivo xml

importamos la librer√≠a **xml.dom.minidom**

```python
import xml.dom.minidom
```

Lo primero de todo parseamos el archivo xml con la funci√≥n `parse`, este procedimiento es similar a utilizar la funci√≥n `open`

```python
raw_xml = xml.dom.minidom.parse('books.xml')
```

```python
raw_xml
```

En casi todos los XML tendremos siempre un Elemento principal que abre y cierra el XML, en nuestro caso es Catalog. Para poder procesarlo utilizaremos la funci√≥n <code>__documentElement__</code>

```python
libros_object = raw_xml.documentElement
```

```python
libros_object
```

Como podemos comprobar, ahora nuestro objeto DOM ha encontrado el elemento __Catalog__, por lo que ahora podremos iterar y procesar todos sus elementos o nodos. En este caso, simplemente tenemos elementos Book bajo el elemento general Catalog. Por lo tanto, utilizaremos la funci√≥n <code>__getElementsByTagName(Nombre_Nodo)__</code>

```python
books = libros_object.getElementsByTagName('Book')
```

```python
books
```

Si atendemos a la lista capturada, podremos ver que tenemos dos objetos DOM en los que el elemento es Book, ahora, ya podemos iterar sobre el resto de elementos de cada libro. Si nos fijamos bien, estamos realizando la misma operaci√≥n de Catalog a Book que de Book a Autor, por lo que de nuevo utilizaremos <code>__getElementsByTagName(Nombre_Nodo)__</code>, en este caso, para ir accediendo a cada campo, de forma iterativa con un bucle for

```python
for item in books:
    Author = item.getElementsByTagName('Author')
    print(Author)
```

Como ya no tenemos m√°s elementos sobre los que iterar, si no que ya queremos acceder directamente al contenido, por ejemplo el primer autor: Garghentini, Davide, tenemos que acceder a los elementos de las listas DOM Element, como podemos comprobar cada lista son de longitud 1, no obstante, lo comprobamos.

```python
for item in books:
    print(len(item.getElementsByTagName('Author')))
#     Author = item.getElementsByTagName('Author')
#     print(Author)
```

Es decir, las listas son indexables por [0], aqu√≠ es importante recordar que si tuvi√©ramos m√°s elementos sobre Author, tendr√≠amos que crear un nuevo bucle for para ir recorriendo estos subelementos. Por lo tanto sobre lista del elemento Author, para acceder al contenido utilizamos la funci√≥n <code>__childNodes__</code> que nos devolver√° una nueva lista

```python
for item in books:
    Author = item.getElementsByTagName('Author')[0].childNodes
    print(Author)
```

De nuevo hemos obtenido una lista por cada autor de cada libro, vamos a ver la longitud de estas listas.

```python
for item in books:
    print(len(item.getElementsByTagName('Author')[0].childNodes))
#     Author = item.getElementsByTagName('Author')[0].childNodes
#     print(Author)
```

De nuevo, tenemos solamente un autor, por lo tanto son indexables por [0]

```python
for item in books:
    Author = item.getElementsByTagName('Author')[0].childNodes[0]
    print(Author)
```

Finalmente, vemos que estamos en un nodo conel atributo Text, para poder recuperar este texto, tenemos que hacer uso de la funci√≥n <code>__.data__</code>

```python
for item in books:
    Author = item.getElementsByTagName('Author')[0].childNodes[0].data
    print(Author)
```

Ya hemos llegado al nombre de cada autor, para comprender este c√≥digo de una forma m√°s sencilla donde en lugar de obtener un √≠tem por l√≠nea, realicemos nuevos bucles por si la longitud de las lista es diferente podr√≠amos reescribir el anterior c√≥digo a lo siguiente.

```python
for item in books:
    target = 'Autor'
    for tag in item.getElementsByTagName('Author'):
        for nodo in tag.childNodes:
            autor = nodo.data
            print(autor)
```

Finalmente, podemos mostrar toda la informaci√≥n del XML iterando sobre los nodos hijos del mismo modo que hemos realizado con el autor.

```python
for item in books:
    print('Procesando libro ...')
    
    # Recuperamos los campos
    Author = item.getElementsByTagName('Author')[0].childNodes[0].data
    Title = item.getElementsByTagName('Title')[0].childNodes[0].data
    Genre = item.getElementsByTagName('Genre')[0].childNodes[0].data
    Price = item.getElementsByTagName('Price')[0].childNodes[0].data
    PublishDate = item.getElementsByTagName('PublishDate')[0].childNodes[0].data
    Description = item.getElementsByTagName('Description')[0].childNodes[0].data
    
    # Imprimimos los campos
    print('\t * Autor: {} '.format(Author))
    print('\t * T√≠tulo: {} '.format(Title))
    print('\t * G√©nero: {} '.format(Genre))
    print('\t * Precio: {} '.format(Price))
    print('\t * Fecha publicaci√≥n: {} '.format(PublishDate))
    print('\t * Descripci√≥n: {} '.format(Description))
    print('-'*25)
```

An√°logamente utilizando m√°s bucles para los √≠ndices ...

```python
for item in books:
    print('Procesando libro ...')
    
    target = ['Autor', 'Title', 'Genre', 'Price', 'PublishDate', 'Description']
    
    for elemento in target:
        # Iteramos por cada elemento
        for tag in item.getElementsByTagName(elemento):
            # Iteramos por cada tag name
            for nodo in tag.childNodes:
                # Iteramos por cada nodo
                final_data = nodo.data
                print('\t * {}: {} '.format(elemento, final_data))
    print('-'*25)
```
