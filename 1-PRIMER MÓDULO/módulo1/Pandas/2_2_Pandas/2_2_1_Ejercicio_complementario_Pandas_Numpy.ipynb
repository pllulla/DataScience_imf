## 1

Lee el archivo como un dataframe a través de pandas.

```python
import pandas as pd

df = pd.read_csv("Spotify-2000.csv")
```

```python
df.head(6)
```

## 2

Ordena el dataframe por la columna 'Popularity' y muestra el título, artista y año de las 10 canciones más populares

```python
df[['Title', 'Artist', 'Year', 'Popularity']].sort_values(['Popularity'], ascending=False)
```

```python
df[['Title', 'Artist', 'Year', 'Popularity']].sort_values(['Popularity'], ascending=False).head(10)
```

## 3

Muestra la matriz de correlación entre las columnas BPM, Energy y Danceability

```python
df.columns
```

```python
df[['Beats Per Minute (BPM)', 'Energy', 'Danceability']].corr()
```

## 4

¿Cuántas canciones hay por cada año?, agrupa el dataframe por la columna year y muestra el conteo de canciones

```python
df.groupby(['Year']).count()['Index']
```

## 5

Convierte en un array de Numpy la columna Popularity y, obtén el número de fila que contiene la canción menos popular, posteriormente, muestra la fila completa con la canción menos popular en el dataframe.

```python
import numpy as np

popularidad = np.array(df['Popularity'])
```

```python
menos_popular = popularidad.argmin()
```

```python
menos_popular
```

```python
df[menos_popular:menos_popular+1]
```
