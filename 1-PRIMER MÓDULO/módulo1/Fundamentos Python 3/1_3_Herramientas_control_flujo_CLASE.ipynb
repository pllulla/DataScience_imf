### Ejercicio 12.

Escribe un programa que solicite un número entero mayor que cero y cualcule su factorial.

NOTA: La función factorial se representa con un signo de exclamación “!” detrás de un número. Esta exclamación quiere decir que hay que multiplicar todos los números enteros positivos que hay entre ese número y el 1.

6! = 1\*2\*3\*4\*5\*6 = 720

#### Solución

```python
numero = int(input("Escribe un número entero positivo: "))
if numero <= 0 :
    print("¡Debías escribir un número entero positivo!")
else:
    factorial = 1
    for i in range(1, numero + 1):
        print("\tValor de factorial {} próximo valor de factorial {}".format(factorial, factorial * i))
        factorial = factorial * i
    print("El factorial de ",numero," es ",factorial)
```

### Ejercicio 18. 

Variación del programa anterior, en este caso, se preguntará al usuario cuántas sumas son correctas para superar el reto, una vez finalizado el reto, se mostrará el número de intentos empleados, entendiendo como intento tanto si el usuario acierta como si falla la suma.

```python
import random

print("Escribe el resultado de las siguientes operaciones: ")
sumas_correctas = 0
```

```python
n_correctas = int(input("¿Cuántas sumas son correctas? "))
```

```python
aciertos = 0
fallos   = 0

while sumas_correctas < n_correctas:
    
    x = random.randrange(1, 21)
    y = random.randrange(1, 21)
    
    respuesta = int(input(f"{x} + {y} = "))
    
    if respuesta == x + y:
        sumas_correctas += 1
        print("¡Correcto!")
        aciertos = aciertos + 1
        
    else:
        print("¡Incorrecto!")
        fallos = fallos + 1
        

print("Reto superado.")
print("\tNúmero de intentos empleados {}".format(aciertos + fallos))
```
